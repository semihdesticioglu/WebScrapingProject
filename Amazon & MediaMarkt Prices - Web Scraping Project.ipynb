```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import time

## Define a function to run the url. If it fails, it will try up to 10 times and then will give a failure warning.
def process(try_number=0):
    try:
        driver.get(url)
        time.sleep(10)
        return 0
    except:
        time.sleep(10)
        driver.quit()
        if try_number > 10:
            print('{} kere hata almistir'.format(try_number))
        else:
            return process(try_number+1)


agent = {"User-Agent":'Mozilla/5.0 (X11; Ubuntu; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36 RuxitSynthetic/1.0 v5214140143 t55095'}
path = r'C://Users//TKA//geckodriver.exe'
opts = Options()
opts.add_argument("Mozilla/5.0 (X11; Ubuntu; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36 RuxitSynthetic/1.0 v5214140143 t55095")
driver = webdriver.Firefox(executable_path = path, options=opts)

#########################################  AMAZON SMARTPHONE ########################################################

## Create lists for saving product models, brands, old prices and the prices.
modeller = []
markalar = []
eskifiyatlar = []
fiyatlar = []

## We will create a check variable that is false. It will turn true when "next page" icon displayed.
## Then it will stop checking the next page.
check = False
while not check:
    for i in range(1,55):
        url = "https://www.amazon.com.tr/s?i=electronics&bbn=13709907031&rh=n%3A12466496031%2Cn%3A12466497031%2Cn%3A13709880031%2Cn%3A13709907031%2Cp_6%3AA1UNQM1SR2CHM&dc&page=" + str(i) + "&fst=as%3Aoff&qid=1597309002&rnid=15358539031&ref=sr_pg_" + str(i)
        process()
        html = driver.execute_script("return document.documentElement.outerHTML")
        sel_soup = BeautifulSoup(html, "html.parser")
        for i in range(1,len(sel_soup.find_all("div",{"class":"a-section a-spacing-medium"}))+1):
            urun =sel_soup.find_all("div",{"class":"a-section a-spacing-medium"})[i-1]
            model = urun.find_all("span",{"class":"a-size-base-plus a-color-base a-text-normal"})[0].text
            marka = urun.find_all("span",{"class":"a-size-base-plus a-color-base a-text-normal"})[0].text.split(" ")[0]
            markalar.append(marka)
            modeller.append(model)
            if len(urun.find_all("span",{"class":"a-price-whole"})) != 0:
                fiyat = urun.find_all("span",{"class":"a-price-whole"})[0].text.strip(",")
                fiyatlar.append(fiyat)
            else:
                fiyatlar.append("")
            if len(urun.find_all("span",{"class":"a-offscreen"})) != 0:  
                eskifiyat = urun.find_all("span",{"class":"a-offscreen"})[-1].text
                eskifiyatlar.append(eskifiyat)
            else:
                eskifiyatlar.append("")
        if len(sel_soup.find_all("li",{"class":"a-disabled a-last"})) != 0:
            check = True
            break
## Define a dictionary from lists we used and create a dataframe with this dictionary.
dict = {"marka":markalar,"model":modeller,"eski fiyat":eskifiyatlar,"fiyat":fiyatlar}
dfAmazon_Tel = pd.DataFrame(dict)
print("Amazon tel ok")


#########################################  AMAZON HIT LIST ########################################################

## Create lists for saving product models, brands, old prices and the prices.
modeller = []
markalar = []
sıralama= []
fiyatlar = []

## We will create a check variable that is false. It will turn true when "next page" icon displayed.
## Then it will stop checking the next page.
check = False
while not check:
    for i in range(1,55):
        url = "https://www.amazon.com.tr/gp/bestsellers/electronics/ref=zg_bs_pg_2/261-6659901-0942336?ie=UTF8&pg=" + str(i)
        process()
        html = driver.execute_script("return document.documentElement.outerHTML")
        sel_soup = BeautifulSoup(html, "html.parser")
        for i in range(1,len(sel_soup.find_all("span",{"class":"a-list-item"}))+1):
            urun = sel_soup.find_all("span",{"class":"a-list-item"})[i-1]
            sıra = urun.find_all("span",{"class":"zg-badge-text"})[0].text
            model = urun.find_all("div",{"class":"p13n-sc-truncated"})[0].text
            marka = urun.find_all("div",{"class":"p13n-sc-truncated"})[0].text.split(" ")[0]
            markalar.append(marka)
            modeller.append(model)
            if len(urun.find_all("span",{"class":"p13n-sc-price"})) == 2:
                fiyat = ""
                fiyat += urun.find_all("span",{"class":"p13n-sc-price"})[0].text
                fiyat += "-"
                fiyat += urun.find_all("span",{"class":"p13n-sc-price"})[0].text
            elif len(urun.find_all("span",{"class":"p13n-sc-price"})) == 1:
                fiyat = urun.find_all("span",{"class":"p13n-sc-price"})[0].text
            else:
                fiyat = ""
            fiyatlar.append(fiyat)
            sıralama.append(sıra)
        if len(sel_soup.find_all("li",{"class":"a-disabled a-last"})) != 0:
            check = True
            break

## Define a dictionary from lists we used and create a dataframe with this dictionary.
dict = {"sıralama":sıralama,"marka":markalar,"model":modeller,"fiyat":fiyatlar}
dfAmazon_Hits = pd.DataFrame(dict)
print("Amazon Top100 ok")



#########################################  MEDIA MARKT SMARTPHONE ########################################################

## Create lists for saving product models, brands, old prices and the prices.
modeller = []
markalar = []
eskifiyat = []
fiyat = []

## We will create a check variable that is false. It will turn true when "next page" icon displayed.
## Then it will stop checking the next page.
check = False
while not check:
    for i in range(1,55):
        url = "https://www.mediamarkt.com.tr/tr/category/_cep-telefonlar%C4%B1-504171.html?searchParams=&sort=&view=&page=" + str(i)
        process()
        html = driver.execute_script("return document.documentElement.outerHTML")
        sel_soup = BeautifulSoup(html, "html.parser")
        ## Find product boxes with "product-wrapper" class and find model, brand and price in it.
        for i in range(0,len(sel_soup.find_all("div",{"class":"product-wrapper"}))):
            urun = sel_soup.find_all("div",{"class":"product-wrapper"})[i]
            modeller.append(urun.find_all("a")[-1].text.strip())
            markalar.append(urun.find_all("a")[-1].text.strip().split(" ")[0])
            price = ""
            fiyatlar = urun.find_all("div", ["price small length-1","price small length-2","price small length-3","price small length-4","price small length-5","price small length-6","price small length-7","price small length-8"])
            if len(fiyatlar) != 0:
                for k in range(1,len(fiyatlar[0].find_all("span"))):
                    price += fiyatlar[0].find_all("span")[k-1].text   
            fiyat.append(price)
            fiyatlar = urun.find_all("div",["price price-xs price-old length-1","price price-xs price-old length-2","price price-xs price-old length-3","price price-xs price-old length-4","price price-xs price-old length-5","price price-xs price-old length-6","price price-xs price-old length-7","price price-xs price-old length-8","price price-xs price-old length-9","price price-xs price-old length-10"])
            price2 = ""
            if len(fiyatlar) != 0:
                for z in range(1,len(fiyatlar[0].find_all("span"))):
                    price2 += fiyatlar[0].find_all("span")[z-1].text   
            eskifiyat.append(price2)
        if len(sel_soup.find_all("li",{"class":"pagination-next"})) == 0:
            check = True
            break

## Define a dictionary from lists we used and create a dataframe with this dictionary.            
dict = {"marka":markalar,"model":modeller,"eski fiyat":eskifiyat,"fiyat":fiyat}
dfMM_Tel = pd.DataFrame(dict)
dfMM_Tel = dfMM_Tel[~dfMM_Tel["model"].str.contains("Outlet")]
dfMM_Tel = dfMM_Tel.reset_index(drop=True)
print("mediamarkt tel ok")


#########################################  MEDIA MARKT TABLET&PC ########################################################

modeller = []
markalar = []
eskifiyat = []
fiyat = []
check = False
while not check:
    for i in range(1,55):
        url = "https://www.mediamarkt.com.tr/tr/category/_ta%C5%9F%C4%B1nabilir-bilgisayarlar-504926.html?searchParams=&sort=&view=&page=" + str(i)
        process()
        html = driver.execute_script("return document.documentElement.outerHTML")
        sel_soup = BeautifulSoup(html, "html.parser")
        ## Find product boxes with "product-wrapper" class and find model, brand and price in it.
        for i in range(0,len(sel_soup.find_all("div",{"class":"product-wrapper"}))):
            urun = sel_soup.find_all("div",{"class":"product-wrapper"})[i]
            modeller.append(urun.find_all("a")[-1].text.strip())
            markalar.append(urun.find_all("a")[-1].text.strip().split(" ")[0])
            price = ""
            fiyatlar = urun.find_all("div", ["price small length-1","price small length-2","price small length-3","price small length-4","price small length-5","price small length-6","price small length-7","price small length-8"])
            if len(fiyatlar) != 0:
                for k in range(1,len(fiyatlar[0].find_all("span"))):
                    price += fiyatlar[0].find_all("span")[k-1].text   
            fiyat.append(price)
            fiyatlar = urun.find_all("div",["price price-xs price-old length-1","price price-xs price-old length-2","price price-xs price-old length-3","price price-xs price-old length-4","price price-xs price-old length-5","price price-xs price-old length-6","price price-xs price-old length-7","price price-xs price-old length-8","price price-xs price-old length-9","price price-xs price-old length-10"])
            price2 = ""
            if len(fiyatlar) != 0:
                for z in range(1,len(fiyatlar[0].find_all("span"))):
                    price2 += fiyatlar[0].find_all("span")[z-1].text   
            eskifiyat.append(price2)
        if len(sel_soup.find_all("li",{"class":"pagination-next"})) == 0:
            check = True
            break

## Define a dictionary from lists we used and create a dataframe with this dictionary.
dict = {"marka":markalar,"model":modeller,"eski fiyat":eskifiyat,"fiyat":fiyat}
dfMM_Tablet = pd.DataFrame(dict)
dfMM_Tablet = dfMM_Tablet[~dfMM_Tablet["model"].str.contains("Outlet")]
dfMM_Tablet = dfMM_Tablet.reset_index(drop=True)
print("mediamarkt tabletpc ok")
driver.quit()
```
